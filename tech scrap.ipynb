```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from matplotlib.dates import DateFormatter
import datetime
import numpy as np
```

```python
source = requests.get('https://www.tigermobiles.com/evolution/#footer').text
soup = BeautifulSoup(source, 'lxml')
```

```python
date = []
summary = []
for div in soup.find_all(class_="timeline-description"):
    date.append(div.find('h4').text)
    summary.append(div.find('p').text)
```

```python
print(date, summary)
```

```python
tech_data = pd.DataFrame(
    {'Date': date,
     'Summary': summary,
    })

tech_data
```

```python
tech_data.drop(index=tech_data.index[:24], 
        axis=0, 
        inplace=True)
```

```python
tech_data.reset_index(drop=True)
```

```python
tech_dates = ['2011-13-02', '2012-01-10', '2013-20-09', '2014-01-08', '2015-15-06','2017-17-09', '2018-06-07', '2019-10-09', '2020-06-03']
tech_dates = [datetime.datetime.strptime(x, '%Y-%d-%m') for x in tech_dates]
tech_keywords = ['AMOLED Display', '4G LTE', 'Fingerprint Scanner', 'Bezel-less Display', 'Android Competition','Facial Recognition', 'Prolonged Battery', 'High-Res Cameras', '5G']
```

```python
pol_source = requests.get('https://www.cfr.org/timeline/us-relations-china').text
pol_soup = BeautifulSoup(pol_source, 'lxml')
```

```python
#print(pol_soup.prettify())
pol_date = []
pol_title = []
for x in pol_soup.find_all('div', class_="timeline-event-container"):
    pol_date.append(x.find('time','content-dates').text)
    pol_title.append(x.find('div','content-headline').text)
    
print(pol_date,pol_title)
```

```python
content = []

for y in pol_soup.find_all('div', class_="content-text"):
    content.append(y.find('p').text)
```

```python
pol_data = pd.DataFrame(
    {'Date': pol_date,
     'Title': pol_title,
     'Content': content
    })

pol_data.drop(index=pol_data.index[:20], 
        axis=0, 
        inplace=True)
pol_data
```

```python
pol_data.reset_index(drop=True)
```
